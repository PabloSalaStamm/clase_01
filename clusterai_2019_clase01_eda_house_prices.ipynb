## Missing data (Garbage in Garbage Out)

![Axis diagram](http://alison.dbsdataprojects.com/wp-content/uploads/sites/82/2016/04/cartoon-metadata.png)

El primer paso que debemos realizar SIEMPRE es conocer la calidad de datos con la que contamos. Esto va a significar los cimientos de nuestro trabajo. Si son de mala calidad eventualmente nuestro trabajo no va a ser bueno. 

Algunos tips que nos sirven para entender la situacion:

1- Hay datos nulos/erroneos? Cuantos? Cual es la proporcion?

2- Existe algun tipo de patron entre los datos que tenemos?

3 - Siempre tratar de entender el contexto y el signigicado de cada tipo de datos que tenemos en el problema que queremos resolver.

Funciones a utilizar:

- isnull : Detecta NaN or NULL, devolviendo TRUE(si es nulo) or FALSE
- sum() : suma de valores, al no especificar suma los TRUE
- sort_values(): ordena los valores de forma ascendiente por default
- count(): conteo de valores, al no especificar cuenta los TRUE
- pd.concat(): unimos dos o mas objetos pandas(dataframes), 
    - axis indica si es de manera columnar o uno abajo del otro, 
    - keys utiliza los valores declarados para formar un eje.

```python
# Evaluamos que columnas (features) tienen mas datos nulos
# Utilizamos varias funciones de pandas: isnull, sum, sort_values, count, concat y head
total = df_train.isnull().sum().sort_values(ascending=False)
# En la variable total aplicamos diferentes funciones a la variable df_train de manera concatenada
# Otra opcion podria haber sido aplicarlas una a una y guardarlas en distintas variables. 
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(30)
```

```python
#Que hacemos con los valores nulos
# Los eliminamos? Cada valor? O la columna entera?
df_train = df_train.drop((missing_data[missing_data['Total'] > 1]).index,1) # eliminamos las columnas 
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index) # eliminamos el registro
df_train.isnull().sum().max() #chequeamos que no haya otro valor nulo haciendo el maximo de la suma de cada columna
```

### Data standarization

Existen muchos tipo de estandarizacion o normalizacion de datos, pero su objetivo es ajustar los valores medidos a una escala comun. 
Los metodos mas comunes son: 
- Convertir los datos para que tengan una media cero y un desvio estandard 1. 
- Convertir los datos para que tengan valores entre cero y uno. 

```python
#standardizing data
saleprice_scaled = StandardScaler().fit_transform(df_train['SalePrice'][:,np.newaxis]); 
#StandardScaler usa la media y el desvio estandard para calcular el valor final
# np.newaxis lo usamos para darle la dimension columna a nuestor datos
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10] 
# argsort me devuelve los indices del array al cual estoy aplicando la funcion
high_range= saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```

Por que hicimos esto? De que nos sirve?

 - Muchos test y metodos estadisticos se basan en la normalidad de las variables 
 - Podemos evitar futuros problemas (ej. heteroscedacity: diferente rango de varianza a lo largo de sus variables predictoras) http://www.statsmakemecry.com/smmctheblog/confusing-stats-terms-explained-heteroscedasticity-heteroske.html
 - Importante: De esta manera no aseguramos normalidad con respecto a todas las variables!
 - El analisis puede ser mucho mas completo pero por ahora lo dejamos aca.

```python
#histograma y grafico de probabilidad normal
sns.distplot(df_train['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

Es una especie de distribucion normal pero tiene un corrimiento y esta muy concentrada...

Pero no todo esta perdido, en caso de un desvio positivo podemos aplicar una transformacion logaritimica, es decir un cambio de escala de los datos. 

```python
# Dos amigos Skewness y Kurtosis
print("Skewness: %f" % df_train['SalePrice'].skew())
print("Kurtosis: %f" % df_train['SalePrice'].kurt())
```

```python
#transformacion logaritmica
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```python
# y como un truco de magia...
sns.distplot(df_train['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```
